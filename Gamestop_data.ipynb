```python
import requests
```

```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/stock.html"
```

```python
html_data = requests.get(url).text
print (html_data)
```

```python
from bs4 import BeautifulSoup
```

```python
# Create a BeautifulSoup object
soup = BeautifulSoup(html_data, 'html.parser')
```

```python
soup.find_all("tbody")[1]
```

```python
import pandas as pd
```

```python
# Loop through each row in the table
for row in soup.find_all('tbody')[1].find_all('tr'):
    # Create a list to store the row data
    row_data = []
```

```python
# Loop through each row in the table
for row in soup.find_all('tbody')[1].find_all('tr'):
    # Create a list to store the row data
    row_data = []
    # Loop through each cell in the row
    for cell in row.find_all('td'):
        # Append the cell data to the row_data list
        row_data.append(cell.text)
    # Print the row data
    print(row_data)
```

```python
import matplotlib.pyplot as plt

# Assuming you have stored the data in a list called `row_data`

# Extract x and y values from the data
x = [row[0] for row in row_data]
y = [row[1] for row in row_data]

# Create a line graph
plt.plot(x, y)

# Add labels and title
plt.xlabel('Revenue')
plt.ylabel('Date')
plt.title('GameStop Stock')

# Display the graph
plt.show()
```
